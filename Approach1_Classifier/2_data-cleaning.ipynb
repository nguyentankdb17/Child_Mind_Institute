## Visualize how many NaN values are there in the training dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count the number of missing values in each feature
missing_values = train_df.isnull().sum()

# Plotting the chart
plt.figure(figsize=(12, 30))
ax = missing_values.plot(kind='barh', color='skyblue')

for index, value in enumerate(missing_values):
    ax.text(value, index, str(value), va='center', ha='left', color='black', fontweight='bold')

plt.title('Number of Missing Values in Each Feature')
plt.ylabel('Features')
plt.xlabel('Number of Missing Values')
plt.show()
```
